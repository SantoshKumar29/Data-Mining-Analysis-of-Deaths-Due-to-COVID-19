```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\ksant\Downloads\Compressed\Data Mining Analysis of Deaths Due to COVID-19\Code\outputs\covid_deaths_cleaned.csv")
df["date"] = pd.to_datetime(df["date"])
```

Trend Over Time (line)

```python
trend = df.groupby("date").size()

plt.figure(figsize=(12,4))
plt.plot(trend.index, trend.values)
plt.title("Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```

Deaths by State (bar)

```python
state_counts = df["state"].value_counts()

plt.figure(figsize=(10,5))
state_counts.plot(kind="bar")
plt.title("Deaths by State")
plt.xlabel("State")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```

Deaths by Age Group (bar)

```python
age_grp = df["age_group"].value_counts().sort_index()

plt.figure(figsize=(8,4))
age_grp.plot(kind="bar")
plt.title("Deaths by Age Group")
plt.xlabel("Age group")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```

Deaths by Gender (bar)

```python
df["gender"] = df["male"].map({1:"Male", 0:"Female"})
gender_counts = df["gender"].value_counts()

plt.figure(figsize=(5,4))
gender_counts.plot(kind="bar")
plt.title("Deaths by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```

Comorbidity vs No Comorbidity

```python
df["comorb_label"] = df["comorb"].map({1:"With comorbidity", 0:"No comorbidity"})
comorb_counts = df["comorb_label"].value_counts()

plt.figure(figsize=(6,4))
comorb_counts.plot(kind="bar")
plt.title("Comorbidity Status")
plt.xlabel("Group")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```

Vaccination Status

```python
vax_counts = df["vax_status"].value_counts()

plt.figure(figsize=(8,4))
vax_counts.plot(kind="bar")
plt.title("Deaths by Vaccination Status")
plt.xlabel("Vaccination status")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```

 Vaccine Brand (for vaccinated only)

```python
vaccinated = df[df["primary_brand"].notna()]
brand_counts = vaccinated["primary_brand"].value_counts().head(10)

plt.figure(figsize=(10,4))
brand_counts.plot(kind="bar")
plt.title("Deaths by Primary Vaccine Brand (Top 10)")
plt.xlabel("Brand")
plt.ylabel("Number of deaths")
plt.tight_layout()
plt.show()
```
